```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
xlsx = pd.ExcelFile('AMPCapstoneProject2022-1.xlsx')
```

```python
dfb = pd.read_excel(xlsx, 'Buildings')
df1 = pd.read_excel(xlsx, 'Date Range 1')
df2 = pd.read_excel(xlsx, 'Date Range 2')
df3 = pd.read_excel(xlsx, 'Date Range 3')
```

```python
# Buildings
dfb
```

```python
#combine the 3 date range tables together
df = pd.concat([df1, df2, df3])
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.insert(loc=0, column='row_num', value=np.arange(len(df)))
```

```python
df.set_index('row_num', inplace = True)
```

```python
# clean up
# going to drop sample count as it is almost always 12 with very little deviation
df.drop(columns=['Sample Count'], inplace = True)
```

```python
df
```

```python
df.reset_index(inplace = True)
df.drop(columns=['row_num'], inplace=True)
```

```python
min(df['Timestamp'])
```

```python
max(df['Timestamp'])
```

```python
# date range is from April 1st 2019 to March 31 2022
```

```python
df['Day']=df['Timestamp'].map(lambda x: x.day)
df['Month']=df['Timestamp'].map(lambda x: x.month)
df['Year']=df['Timestamp'].map(lambda x: x.year)
```

```python
# group by NAME and Day/Month/Year
```

```python
aggregate_list = ['Name','Point','Year','Month','Day']
```

```python
grouped_df = df.groupby(aggregate_list).agg({'Total': ['mean','max','sum']})
```

```python
grouped_df.columns = ['Total_mean','Total_max', 'Total_sum']
```

```python
grouped_df = grouped_df.reset_index()
```

```python
print(grouped_df)
```

```python
grouped_df.head(50)
```

### I've grouped the data by Building Name, Energy Demand Point and then broke out the timestamp and summarized it by day instead of the hourly point.  The Total Sum column is the sum energy consumption for that day, for that building, from that energy demand point.  I am not sure what the min, max, and latest data points are in the excel so we should ask about that but I am assuming those are just energy demand points at various points through out the hour long timestamp and are just aggregates

```python
sns.lineplot(x = 'Month', y='Total_sum', data = grouped_df)
# Looks like energy consumption goes up in the summer and drops as it gets colder
# possible that AC is the culprit?
```
